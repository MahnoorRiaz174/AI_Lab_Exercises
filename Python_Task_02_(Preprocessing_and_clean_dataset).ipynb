```python
import pandas as pd
numeric_cols = df.select_dtypes(include=['number']).columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = (df[numeric_cols] < lower_bound) | (df[numeric_cols] > upper_bound)
print("Number of outliers per column:\n", outliers.sum())
df_clean = df[~outliers.any(axis=1)].copy()
print(" Data cleaned successfully!")
print("Original rows:", len(df))
print("Rows after removing outliers:", len(df_clean))
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

numeric_cols = df_clean.select_dtypes(include=['number']).columns

scaler = MinMaxScaler()
df_normalized = df_clean.copy()
df_normalized[numeric_cols] = scaler.fit_transform(df_clean[numeric_cols])
print(" Data normalized successfully using MinMax Scaling!")
df_normalized.head()
```

```python
df_clean.to_csv('Cleaned_AI_ML_Salaries.csv', index=False)
print(" Cleaned dataset saved successfully as 'Cleaned_AI_ML_Salaries.csv'")
files.download('Cleaned_AI_ML_Salaries.csv')
```
